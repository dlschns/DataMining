### 2.2 Clustering
Clustering adalah membagi data kedalam grup-grup yang memiliki objek yang karakteristiknya sama. #### 2.2.1 K-Means Clustering
Tugas akhir ini menggunakan metode **K-Means** untuk pengelompokan data COVID-19 dari library **sklearn**. Algoritma K-Means merupakan algoritma pengelompokan iterative yang melakukan partisi set data kedalam jumlah K cluster yang sudah ditetapkan diawal. Algoritma K-Means sederhana untuk diimplementasikan  dan dijalankan, relatif cepat, mudah beradaptasi, dan umum dalam penggunannya dalam praktek. 

Algoritma K-means bertujuan untuk memilih centroid yang meminimalkan kriteria jarak jumlah-kuadrat dalam cluster:



>Beberapa tahapan K-Means yang dilakukan: 
* Menentukan nilai K 
* Memilih random data sebagai centroid, dipilih sebanyak K yang sudah ditentukan di awal
* Tetapkan data lainnya ke setiap centroid cluster terdekat
* Hitung ulang centroid dari cluster yang baru terbentuk
* Ulang tahap 3-4 hingga kondisi stop looping terpenuhi. 

Kondisi berhenti pada k-means yang biasa digunakan adalah ketika centroid dari cluster baru tidak berubah terus-menerus, titik-titik data selalu berada di cluster yang sama terus-menerus, dan mencapai batas iterasi yang sudah ditentukan.

## III. Percobaan

### 3.1 Preprocessing

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
# membaca file data
df = pd.read_excel(r'TEST_COVID_ASEAN.xlsx')
df
```

```python
Data = df[['date','new_cases','new_deaths','total_tests']]
```

#### Data cleaning: eliminasi missing values

```python
Data = pd.DataFrame(Data.dropna())
Data
```

```python
# mengambil kolom jumlah tes dan kematian baru
DataCovid = Data.iloc[:, 2:4]
DataCovid
```

```python
#menampilkan 5 grup data awal
DataCovid.head()
```

```python
# visualisasi persebaran data
ax = sns.regplot(x=DataCovid['new_deaths'], y=DataCovid['total_tests'], data=Data, fit_reg=False)
```

```python
#mengubah data ke bentuk array agar dapat dilakukan clustering
arrayData = np.array(DataCovid)
print(arrayData)
```

#### Data Transformation: Normalisasi data

```python
scaler = MinMaxScaler()
Normalisasi = scaler.fit_transform(arrayData)
Normalisasi
```

### 3.2 K-Means Clustering

```python
# melakukan clustering
kmeans = KMeans(n_clusters = 3, random_state=123)
kmeans.fit(Normalisasi)
```

```python
#Mencetak pusat dari cluster
print(kmeans.cluster_centers_)
```

```python
# Menampilkan hasil kluster
print(kmeans.labels_)
DataCovid["kluster"] = kmeans.labels_
DataCovid.head()
```

### 3.3 Visualisasi Data

```python
fig, ax = plt.subplots()
sct = ax.scatter(Normalisasi[:,1], Normalisasi[:,0], s = 100,
c = DataCovid.kluster, marker = "o", alpha = 0.5)
centers = kmeans.cluster_centers_
ax.scatter(centers[:,1], centers[:,0], c='red', s=200, alpha=0.5)
plt.title("Hasil Klustering K-Means")
plt.xlabel("KEMATIAN")
plt.ylabel("Test Covid")
plt.show()
```
